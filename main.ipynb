```julia
rt = 2
newF = Vector{Tuple{Int,Int,Bool}}[[(1, 1, 0), (3, 4, 1), (3, 6, 1), (1, 3, 0)], [(2, 1, 0), (4, 4, 1), (4, 3, 1), (2, 6, 0)]];
```

```julia
fem = fe_mabel(marge);
```

```julia
unified_mabel = Dict([(l, l) for l in vcat(f_mabel(marge)...)])

for e in findall(E -> E.t == rt, marge.cc.E)
    fs = findall(F -> e in getproperty.(F, :e), marge.cc.F)
    @assert(length(fs) == 2)
    for (l1, l2) in zip(fem[(fs[1], e)], fem[(fs[2], e)])
        unified_mabel[l2] = l1
    end
end

a = []
for nF in newF
    push!(a, [])
    for (f, e, s) in nF
        append!(a[end], [unified_mabel[l] for l in reversebool(fem[(f, e)], s)])
    end
end
# a
# mabel_dyck.(remabel.(a))
unused = setdiff(values(unified_mabel), vcat(a...))
unused_ground = [m + 1 for m in unused]
ng = sum(marge.n) + length(marge.cc.F)
used_ground = setdiff(1:ng, unused_ground)
```

```julia
gmat = groundmatrix(marge)
nt = length(marge.cc.T) + 1
unused_neighbor = [findall(==(nt), gmat[ug,:]) for ug in unused_ground]
```

```julia
flforest = floatingforest(marge)
for ug in unused_ground
    for un in findall(==(nt), gmat[ug,:])
        add_edge!(flforest, ug, un)
    end
end
```

```julia
T = []
for g in used_ground
    push!(T, rooted_tree_to_dyck(flforest, g))
end
T
```
